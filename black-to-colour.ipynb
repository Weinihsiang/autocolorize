```python
from PIL import Image
import os

def convert_images_to_grayscale(input_dir, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)  # Create the output directory if it doesn't exist

    for filename in os.listdir(input_dir):
        if filename.endswith(('.jpg', '.JPEG', '.png')):  # Process only image files
            input_path = os.path.join(input_dir, filename)
            output_path = os.path.join(output_dir, filename)

            image = Image.open(input_path)
            grayscale_image = image.convert("L")
            grayscale_image.save(output_path)

# Paths to input and output directories for training and validation images
input_train_dir = '/kaggle/input/imagenet/imagenet/train'
input_val_dir = '/kaggle/input/imagenet/imagenet/val'
output_train_dir = 'output_train'
output_val_dir = 'output_val'

# Convert training images to grayscale and save in output_train directory
convert_images_to_grayscale(input_train_dir, output_train_dir)

# Convert validation images to grayscale and save in output_val directory
convert_images_to_grayscale(input_val_dir, output_val_dir)
```

```python
import os

def count_images_in_directory(directory):
    count = 0
    for filename in os.listdir(directory):
        if filename.endswith(('.jpg', '.jpeg', '.png')):  # Consider only image files
            count += 1
    return count

# Paths to output directories
output_train_dir = 'output_train'
output_val_dir = 'output_val'

# Count images in output_train and output_val directories
train_images_count = count_images_in_directory(output_train_dir)
val_images_count = count_images_in_directory(output_val_dir)

print(f"Number of images in 'output_train': {train_images_count}")
print(f"Number of images in 'output_val': {val_images_count}")
```
